```python
Q1. Difference between for loop and while loop with examples

For Loop: Use when you know the number of iterations in advance or when you want to iterate over a sequence (like a list, range, or string).

Example:

# Printing numbers from 1 to 5 using a for loop
for i in range(1, 6):
    print(i)

Here, we know we want to print numbers from 1 to 5, so we use a for loop.

While Loop: Use when the number of iterations isn't known in advance, but it depends on a condition being true.

Example:

# Printing numbers from 1 to 5 using a while loop
i = 1
while i <= 5:
    print(i)
    i += 1

Here, the loop will run until the condition i <= 5 becomes false.



---

Q2. Python Program to Calculate Sum and Product of the First 10 Natural Numbers

# Using for loop
sum_for = 0
product_for = 1
for i in range(1, 11):
    sum_for += i
    product_for *= i

print("Using for loop - Sum:", sum_for)
print("Using for loop - Product:", product_for)

# Using while loop
sum_while = 0
product_while = 1
i = 1
while i <= 10:
    sum_while += i
    product_while *= i
    i += 1

print("Using while loop - Sum:", sum_while)
print("Using while loop - Product:", product_while)


---

Q3. Python Program to Compute the Electricity Bill for a Household

# Taking user input for electricity units
units = int(input("Enter the units of electricity consumed: "))
bill = 0

if units <= 100:
    bill = units * 4.5
elif units <= 200:
    bill = (100 * 4.5) + ((units - 100) * 6)
elif units <= 300:
    bill = (100 * 4.5) + (100 * 6) + ((units - 200) * 10)
else:
    bill = (100 * 4.5) + (100 * 6) + (100 * 10) + ((units - 300) * 20)

print("The total electricity bill is:", bill)

Explanation:

For the first 100 units, we charge Rs. 4.5 per unit.

For the next 100 units (101-200), Rs. 6 per unit.

For the next 100 units (201-300), Rs. 10 per unit.

For units above 300, Rs. 20 per unit.



---

Q4. Program to Calculate Cube and Filter Based on Divisibility by 4 or 5

# Creating a list of numbers from 1 to 100
numbers = list(range(1, 101))
filtered_numbers_for = []
filtered_numbers_while = []

# Using for loop
for num in numbers:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        filtered_numbers_for.append(num)

print("Filtered numbers using for loop:", filtered_numbers_for)

# Using while loop
i = 0
while i < len(numbers):
    cube = numbers[i] ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        filtered_numbers_while.append(numbers[i])
    i += 1

print("Filtered numbers using while loop:", filtered_numbers_while)


---

Q5. Program to Count Vowels in a Given String

# Given string
string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
count = 0

for char in string:
    if char in vowels:
        count += 1

print("Number of vowels in the string:", count)

In this code:

We define the vowels string to check against.

We loop through each character in the input string and increment the count if it is a vowel.
```
